# Урок 1. Формирование гипотез и их скоринг на основе данных
## Задание 1. Выберите любой продукт.
Например ваш продукт

* ozon.ru

* citilink.ru

* telegram

Мной выбран продукт – **citilink.ru**

Какие бы ключевые метрики вы выделили? Опишите почему вы выбрали именно их и можно ли их измерить:

Первая ключевая метрика – это **скорость загрузки сайта**, является важнейшей метрикой для определения эффективности сайта. Как правило, быстрый сайт, привлекает больше пользователей, чем медленный, т. к. продукты загружаются быстро. С помощью специальных веб-сервисов, можно измерить скорость загрузки сайта. Например, сервисы из России, такие как:

https://be1.ru/site-speed/

https://seolik.ru/speed

https://sitespeed.ru/ и другие.

Вторая ключевая метрика – **среднее время на сайте**, является показателем эффективности сайта, показывает, сколько времени пользователи затрачивают на сайте. В теории, если пользователь находится на сайте длительное время, тем вероятнее, что он совершит покупку или наоборот, откажется от неё. Данную метрику можно измерить с помощью Google-аналитики или с помощью других инструментов аналитики. 

Третья ключевая метрика – **количество просмотренных страниц** – показатель количества страниц, просматриваемых пользователем на сайте. В теории, если посетитель заинтересован в продукции и готов к покупке – то будет просматривать большое количество страниц. Данную метрику можно измерить с помощью Google-аналитики или с помощью других инструментов аналитики. 

Четвёртая ключевая метрика – **отношение новых посетителей к вернувшимся**. Показывает, сколько новых посетителей приходят на сайт и сколько вернувшихся. Лучше, когда у сайта больше вернувшихся посетителей, что может свидетельствовать об их удовлетворенности продукцией и сервисом. Данную метрику можно измерить с помощью Google-аналитики или с помощью других инструментов аналитики. 

Пятая ключевая метрика – **конверсия** - показатель того, как часто посетители сайта переходят от просмотра купленных товаров. Измерить конверсию можно с помощью уникальных преференциальных ссылок и кодов отслеживания.

Представленные ключевые метрики, позволяют получить полноценный анализ эффективности продукта – citilink.ru

## Задание 2. Сформулируйте 5-10 гипотез для выбранного продукта.
**Первая гипотеза** – добавление новых опций доставки или повышение скорости работы существующих способов доставки могут повысить лояльность клиентов, а также увеличить конверсию.

**Вторая гипотеза** – запуск программы лояльности с бонусами за покупки и участие в акциях может увеличить количество повторных посещений и относительно конверсии.

**Третья гипотеза** – разработка и интеграция на сайте таких пользовательских функций, как фильтры поиска, порекомендованные товары и подобные, может повысить среднюю корзину и увеличить число продаж.

**Четвёртая гипотеза** – улучшение мобильной оптимизации сайта может повысить конверсию и время пребывания на сайте, так как большинство людей использует смартфоны для поиска и покупки товаров.

**Пятая гипотеза** – расширение ассортимента товаров может увеличить среднее время пребывания на сайте и страницу, на что можно обращать внимание на главной странице сайта.
